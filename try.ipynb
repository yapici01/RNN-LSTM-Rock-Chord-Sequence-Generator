```python
import pandas as pd
import mido
groove = pd.read_csv("groove/info.csv")

filtered_groove = groove[(groove['time_signature'] == '4-4') & (groove['beat_type'] == 'beat') & (groove['style'] == 'funk')]

foo = filtered_groove["midi_filename"].iloc[0]

midi = mido.MidiFile(f"groove/{foo}")

for m in midi:
    print(m)
```
